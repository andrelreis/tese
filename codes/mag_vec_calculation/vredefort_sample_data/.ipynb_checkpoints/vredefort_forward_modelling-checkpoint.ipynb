## Loading from a .txt file

```python
files = ['real_data/MAPx250920150931_1.txt']

nfiles = len(files)
```

```python
data = [] #in V
shape_list = []
for i in range(nfiles):
    data.append(360.*(np.loadtxt(files[i]).T))
    shape_list.append(data[i].shape)
    print shape_list[i]    
```

### Parameters of acquisition

```python
L_top = -15.
L_bottom = 15.
```

```python
shape  = shape_list[0] 
```

```python
Nx, Ny = shape_list[0][0],shape_list[0][1]
```

```python
area = [-1700., 0.,0.,2500.]
```

```python
z_obs = -160. + L_top 
```

```python
xp,yp,zp = regular(area,shape,z_obs)
```

```python
Y = yp.reshape(shape)
X = xp.reshape(shape)
```

### Model for vredefort sample

```python
m1 = .65
inc1 = -90.
dec1 = 0.
mag1 = ang2vec(m1,inc1,dec1)
```

```python
m2 = .075
inc2 = -90.
dec2 = 0.
mag2 = ang2vec(m2,inc2,dec2)
```

```python
m3 = .09
inc3 = - 90.
dec3 = 0.
mag3 = ang2vec(m3,inc3,dec3)
```

```python
model = [PolygonalPrism(verts1,L_top,L_bottom,{'magnetization':mag1}),
         PolygonalPrism(verts2,L_top,L_bottom,{'magnetization':mag2}),
         PolygonalPrism(verts1,L_top,L_bottom,{'magnetization':mag3})]
```

## Generating synthetic data

```python
bz = polyprism.bz(yp,xp,zp,model)
```
