### Generating observation coordinates

```python
analysis['area'] = [-1700., 0.,0.,2500.]
```

```python
xp = np.linspace(analysis['area'][0],analysis['area'][1],modeling['Nx'])
```

```python
yp = np.linspace(analysis['area'][2],analysis['area'][3],modeling['Ny'])
```

```python
Y,X = np.meshgrid(yp,xp)
```

## Visualization of 415 mT map

```python
title_font = 20
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(7,8), tight_layout=True)


plt.title('Ha = 415 mT', fontsize = 18)
plt.contourf(Y,X,modeling['obs_data'], 30,cmap='viridis')
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('mT')
plt.xlabel('y ($\mu m$)', fontsize = 16)
plt.ylabel('x ($\mu m$)', fontsize = 16)
plt.ylim(np.max(X),np.min(X))
plt.xlim(np.min(Y),np.max(Y))

file_name = 'figs/observed_data'
plt.savefig(file_name+'.tiff',dpi=200)
saved_files.append(file_name+'.tiff')

plt.savefig(file_name+'.png',dpi=200)
saved_files.append(file_name+'.png')


plt.show()
```

## Visualization of model data

```python
title_font = 20
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(7,8), tight_layout=True)


plt.title('B_z (model)', fontsize = 18)
plt.contourf(Y,X,modeling['pred_data'].reshape(modeling['shape']), 30,cmap='viridis')
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('mT')
plt.xlabel('y ($\mu m$)', fontsize = 16)
plt.ylabel('x ($\mu m$)', fontsize = 16)
plt.ylim(np.max(X),np.min(X))
plt.xlim(np.min(Y),np.max(Y))

file_name = 'figs/synthetic_data'
plt.savefig(file_name+'.tiff',dpi=200)
saved_files.append(file_name+'.tiff')

plt.savefig(file_name+'.png',dpi=200)
saved_files.append(file_name+'.png')


plt.show()
```

## Sample bounds visualization

## Visualization of sample bounds and the model data

```python
title_font = 20
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(12,6), tight_layout=True)

plt.subplot(1,2,1)
plt.title('(a) Sample Bounds',fontsize=title_font)
plt.plot(bounds['poly1_x'], bounds['poly1_y'],color='b',linestyle='-',linewidth=2)
plt.plot(bounds['poly2_x'], bounds['poly2_y'],color='g',linestyle='-',linewidth=2)
plt.plot(bounds['poly3_x'], bounds['poly3_y'],color='r',linestyle='-',linewidth=2)
plt.xlabel('y ($\mu m$)', fontsize = 16)
plt.ylabel('x ($\mu m$)', fontsize = 16)
plt.ylim(np.max(X),np.min(X))
plt.xlim(np.min(Y),np.max(Y))


plt.subplot(1,2,2)
plt.contourf(Y,X,modeling['pred_data'].reshape(modeling['shape']),30, cmap='viridis')
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('mT')
plt.xlabel('y ($\mu m$)', fontsize = 16)
plt.ylabel('x ($\mu m$)', fontsize = 16)
plt.title('(b) Bz (model)', fontsize=title_font)
plt.ylim(np.max(X),np.min(X))
plt.xlim(np.min(Y),np.max(Y))

file_name = 'figs/model_data'
plt.savefig(file_name+'.tiff',dpi=200)
saved_files.append(file_name+'.tiff')

plt.savefig(file_name+'.png',dpi=200)
saved_files.append(file_name+'.png')


plt.show()
```

## Comparison between model data and observed data

```python
title_font = 16
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(15,7), tight_layout=True)

ranges = saturation_factor*np.abs([modeling['obs_data'].max(),modeling['obs_data'].min()]).max()

plt.subplot(1,3,1)
plt.title('(a) Observed data', fontsize=title_font)
plt.contourf(Y,X,modeling['obs_data'].reshape(modeling['shape']),30, vmin=-ranges, vmax=ranges)
plt.plot(bounds['poly1_x'], bounds['poly1_y'],color='b',linestyle='-',linewidth=2)
plt.plot(bounds['poly2_x'], bounds['poly2_y'],color='g',linestyle='-',linewidth=2)
plt.plot(bounds['poly3_x'], bounds['poly3_y'],color='r',linestyle='-',linewidth=2)
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('mT')
plt.xlabel('y ($\mu m$)', fontsize = 16)
plt.ylabel('x ($\mu m$)', fontsize = 16)
plt.ylim(np.max(X),np.min(X))
plt.xlim(np.min(Y),np.max(Y))

plt.subplot(1,3,2)
plt.title('(b) Bz (model)', fontsize=title_font)
plt.contourf(Y,X,modeling['obs_model'].reshape(modeling['shape']),30,vmin=-ranges, vmax=ranges)
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('mT')
plt.plot(bounds['poly1_x'], bounds['poly1_y'],color='b',linestyle='-',linewidth=2)
plt.plot(bounds['poly2_x'], bounds['poly2_y'],color='g',linestyle='-',linewidth=2)
plt.plot(bounds['poly3_x'], bounds['poly3_y'],color='r',linestyle='-',linewidth=2)
plt.xlabel('y ($\mu m$)', fontsize = 16)
plt.ylabel('x ($\mu m$)', fontsize = 16)
plt.ylim(np.max(X),np.min(X))
plt.xlim(np.min(Y),np.max(Y))

plt.subplot(1,3,3)
r_ranges = saturation_factor*np.abs([modeling['residual'].max(),modeling['residual'].min()]).max()
plt.title('(c) Residual ', fontsize=title_font)
plt.contourf(Y,X,modeling['residual'].reshape(modeling['shape']),30,vmin=-r_ranges, vmax=r_ranges)
plt.plot(bounds['poly1_x'], bounds['poly1_y'],color='b',linestyle='-',linewidth=2)
plt.plot(bounds['poly2_x'], bounds['poly2_y'],color='g',linestyle='-',linewidth=2)
plt.plot(bounds['poly3_x'], bounds['poly3_y'],color='r',linestyle='-',linewidth=2)
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('mT')
plt.xlabel('y ($\mu m$)', fontsize = 16)
plt.ylabel('x ($\mu m$)', fontsize = 16)
plt.ylim(np.max(X),np.min(X))
plt.xlim(np.min(Y),np.max(Y))

file_name = 'figs/bounds_observed_model_comparison'
plt.savefig(file_name+'.tiff',dpi=200)
saved_files.append(file_name+'.tiff')

plt.savefig(file_name+'.png',dpi=200)
saved_files.append(file_name+'.png')

plt.show()
```

### Magnetic moment calculation

### For Handle

```python
'Magnetization for handle in A/m: %.2f' % modeling['m1']
```

```python
analysis['dimension1_y'] = 0.08  # in mm
analysis['dimension1_x'] = 0.07  # in mm
```

```python
analysis['S_1'] = analysis['dimension1_y']*analysis['dimension1_x']*1e-6 # in mm
```

```python
analysis['moment_mag1'] = analysis['S_1']* modeling['m1']
```

```python
print analysis['moment_mag1']
```

```python
'Magnetic moment for handle in $A m^2$: %.7f' % analysis['moment_mag1'] 
```

### For Head

```python
'Magnetization for head in A/m: %.2f' % modeling['m2']
```

```python
analysis['dimension2_y'] = 0.045  # in mm
analysis['dimension2_x'] = 0.070  # in mm
```

```python
analysis['S_2'] =  analysis['dimension2_y']*analysis['dimension2_x']*1e-6 # in mm
```

```python
analysis['moment_mag2'] = analysis['S_2']*modeling['m2']
```

```python
print analysis['moment_mag2']
```

```python
'Magnetic moment for head in Am2: %.7f' % analysis['moment_mag2'] 
```
