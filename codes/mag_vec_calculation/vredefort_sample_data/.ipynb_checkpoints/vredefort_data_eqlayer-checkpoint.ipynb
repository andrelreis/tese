## Loading from a .txt file

```python
files = ['real_data/MAPx280620191934.txt']

nfiles = len(files)
```

```python
data = [] #in V
shape_list = []
for i in range(nfiles):
    data.append(360.*(np.loadtxt(files[i]).T))
    shape_list.append(data[i].shape)
    print shape_list[i]    
```

```python
data_set['obs_data'] = data[0]
```

### Parameters of acquisition

```python
data_set['L_top'] = -148.
data_set['L_bottom'] = 148.
```

```python
data_set['shape'] = shape_list[0] 
```

```python
data_set['Nx'], data_set['Ny'] = shape_list[0][0],shape_list[0][1]
```

```python
data_set['area'] = [-36084.,0.,0.,29636.]
```

```python
data_set['z_obs'] = -138. + data_set['L_top'] 
```

```python
xp,yp,zp = regular(data_set['area'],data_set['shape'],data_set['z_obs'])
```

```python
Y = yp.reshape(data_set['shape'])
X = xp.reshape(data_set['shape'])
```

### Visualization of data set of Vredefort sample

```python
title_font = 20
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(8,8), tight_layout=True)

plt.contourf(Y.reshape(data_set['shape']),X.reshape(data_set['shape']),data[0].reshape(data_set['shape']),
             50,cmap='viridis')
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('mT')
plt.xlabel('y (mm)', fontsize = title_font)
plt.ylabel('x (mm)', fontsize = title_font)
plt.title('Bz (Vredefort model)', fontsize=title_font)


#file_name = 'figs/regular/noisy_data_bz_sphere_RM_regular'
#plt.savefig(file_name+'.png',dpi=200)
#saved_files.append(file_name+'.png')

plt.show()
```
