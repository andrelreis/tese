## Calculating the data for Regular grid

### For regular grid

#### Observation area

```python
print 'Area limits: \n x_max = %.1f m \n x_min = %.1f m \n y_max = %.1f m \n y_min = %.1f m' % (regular['area'][1], regular['area'][0],
                                                                                                regular['area'][3],regular['area'][2])
```

#### Grid information

```python
print 'Shape : (%.0f,%.0f)'% regular['shape'] 
print 'Number of data: %.1f' % regular['N']
print 'dx: %.1f m' % regular['dx']
print 'dy: %.1f m ' % regular['dy']
print 'Height: %.1f m' % regular['z_obs']
```

#### Calculation the data

```python
data_set['bz_sphere_RM_reg'] = sphere.bz(regular['x'],regular['y'],regular['z'],
                                          model_sphere_remanent['model'])
```

```python
data_set['bx_sphere_RM_reg'] = sphere.bx(regular['x'],regular['y'],regular['z'],
                                          model_sphere_remanent['model'])
```

```python
data_set['by_sphere_RM_reg'] = sphere.by(regular['x'],regular['y'],regular['z'],
                                          model_sphere_remanent['model'])
```

##### Generating noise for the data set

```python
np.random.seed(seed=40)
```

```python
std_noise = 10.
r = np.random.normal(0.0,std_noise, regular['Nx']*regular['Ny'])
data_set['bz_obs_sphere_RM_reg'] =  data_set['bz_sphere_RM_reg'] + r
```

#### Visualization of bz-component

```python
title_font = 20
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(8,8), tight_layout=True)

ranges = saturation_factor*np.abs([data_set['bz_obs_sphere_RM_reg'].max(), 
                                   data_set['bz_obs_sphere_RM_reg'].min()]).max()

plt.contourf(regular['y'].reshape(regular['shape']), 
             regular['x'].reshape(regular['shape']),
             data_set['bz_obs_sphere_RM_reg'].reshape(regular['shape']), 
             20, cmap='viridis',vmin=-ranges, vmax=ranges)
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('nT')
plt.xlabel('y (m)', fontsize = title_font)
plt.ylabel('x (m)', fontsize = title_font)
plt.title('Bz (RM_reg_grid)', fontsize=title_font)

file_name = 'figs/regular/noisy_data_bz_sphere_RM_regular'
plt.savefig(file_name+'.png',dpi=200)
saved_files.append(file_name+'.png')

plt.show()
```

#### Visualization of bx-component

```python
title_font = 20
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(8,8), tight_layout=True)

ranges = saturation_factor*np.abs([data_set['bx_sphere_RM_reg'].max(), 
                                   data_set['bx_sphere_RM_reg'].min()]).max()

plt.contourf(regular['y'].reshape(regular['shape']), 
             regular['x'].reshape(regular['shape']),
             data_set['bx_sphere_RM_reg'].reshape(regular['shape']), 
             20, cmap='viridis',vmin=-ranges, vmax=ranges)
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('nT')
plt.xlabel('y (m)', fontsize = title_font)
plt.ylabel('x (m)', fontsize = title_font)
plt.title('Bx (RM_reg_grid)', fontsize=title_font)

file_name = 'figs/regular/noisy_data_bx_sphere_RM_regular'
plt.savefig(file_name+'.png',dpi=200)
saved_files.append(file_name+'.png')

plt.show()
```

#### Visualization of by-component

```python
title_font = 20
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(8,8), tight_layout=True)

ranges = saturation_factor*np.abs([data_set['by_sphere_RM_reg'].max(), 
                                   data_set['by_sphere_RM_reg'].min()]).max()

plt.contourf(regular['y'].reshape(regular['shape']), 
             regular['x'].reshape(regular['shape']),
             data_set['by_sphere_RM_reg'].reshape(regular['shape']), 
             20, cmap='viridis',vmin=-ranges, vmax=ranges)
plt.colorbar(pad=0.01, aspect=40, shrink=1.0).set_label('nT')
plt.xlabel('y (m)', fontsize = title_font)
plt.ylabel('x (m)', fontsize = title_font)
plt.title('By (RM_reg_grid)', fontsize=title_font)

file_name = 'figs/regular/noisy_data_by_sphere_RM_regular'
plt.savefig(file_name+'.png',dpi=200)
saved_files.append(file_name+'.png')

plt.show()
```

#### Generating .pickle file

```python
now = datetime.datetime.utcnow().strftime('%d %B %Y %H:%M:%S UTC')
data_set['metadata'] = 'Generated by {name} on {date}'.format(date=now, name=notebook_name)
```

```python
file_name = 'data/data_set.pickle'
with open(file_name, 'w') as f:
    pickle.dump(data_set, f)
    
saved_files.append(file_name)
```
